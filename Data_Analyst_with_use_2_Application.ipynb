```python
print(soup.find("span",class_='phone').text)
```

```python
Features_testimonials=soup.find_all("div",class_="quote")
for testimonial in Features_testimonials:
    print(testimonial.text)
```

```python
staffMember=soup.find_all("div",class_="info col-xs-8 col-xs-offset-2 col-sm-7 col-sm-offset-0 col-md-6 col-lg-8")
for staff in staffMember:
    print(staff.text)
```

```python
links=soup.find_all("a")
for link in links:
    print(link.text,link.get("href"))
```
